### Q9

그리고 이 tabel을 이용하여 각 월별로 댓글 수를 확인할 수 있는 시각화 자료를 만드세요

```python
data_count = data.groupby(['year','month']).count()
```

```python
data_count
```

```python
data_count_comment = data_count['comment']
```

```python
data_count_comment
```

```python
data_count_comment.plot(title='Youtube comment in 3 years')
```

```python
import matplotlib.pyplot as plt
```

```python
temp = data_count_comment.reset_index()
```

```python
temp
```

```python
def YearAndMonth(x):
    return str(x['year']) + '/' + str(x['month'])
```

```python
lst = temp.apply(lambda x : YearAndMonth(x), axis =1)
```

```python
temp['YM'] = lst
```

```python
YearAndMonth(temp.iloc[0])
```

```python
temp
```

```python
plt.plot(temp['YM'], temp['comment'])
plt.title('comment count')
plt.grid()
plt.legend('Comments')
plt.legend(fontsize=30)
plt.style.use('seaborn-white')

plt.rcParams["figure.figsize"] = (30,10)
```
