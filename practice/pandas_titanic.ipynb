6. Survived는 사고로 살아남은 사람과 죽은 사람을 나타내는 값입니다.

Survived가 1인 사람들과 Survived가 0인 사람들로 group을 나눠서 각각의 age의 평균을 구하세요

Tip : pandas의 아직 배우지 않은 groupby라는 함수를 사용하면 아주 쉽게 수행할 수 있지만

어렵다면 그냥 할 수 있는 방법을 찾아서 해보세요

```python
Sur_MeanAge = data3.groupby('Survived')['Age'].mean()
```
